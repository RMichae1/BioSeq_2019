```julia
initial_state = [0.6, 0.4] # make it more intuitive
# get initial state from the maximum likelihood fit
path = [0.7 0.3; 0.4 0.6]

D = [Categorical([0.5, 0.4, 0.1]), Categorical([0.1, 0.3, 0.6])]

hmm = HMM(initial_state, path, D)

# sample over generated HMM with 250 steps
z, way = rand(hmm, 250)
```
